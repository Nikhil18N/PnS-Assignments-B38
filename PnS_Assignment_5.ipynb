```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

def generate_call_data(N, min_calls=0, max_calls=7):
    return np.random.randint(min_calls, max_calls + 1, size=N)

def compute_pmf(data):
    counts = Counter(data)
    total = sum(counts.values())
    pmf = {k: v / total for k, v in counts.items()}
    return pmf

def compute_cdf(pmf):
    sorted_keys = sorted(pmf.keys())
    cdf = {}
    cumulative_prob = 0
    for k in sorted_keys:
        cumulative_prob += pmf[k]
        cdf[k] = cumulative_prob
    return cdf

def plot_pmf_cdf(pmf, cdf):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.bar(pmf.keys(), pmf.values(), width=0.6, alpha=0.7, color='b', edgecolor='black')
    plt.xlabel("Number of Calls")
    plt.ylabel("PMF")
    plt.title("Probability Mass Function")

    plt.subplot(1, 2, 2)
    plt.plot(list(cdf.keys()), list(cdf.values()), marker='o', linestyle='-', color='r')
    plt.xlabel("Number of Calls")
    plt.ylabel("CDF")
    plt.title("Cumulative Distribution Function")

    plt.show()

# Generate dataset
N = 1000
data = generate_call_data(N)

# Compute PMF and CDF
pmf = compute_pmf(data)
cdf = compute_cdf(pmf)

# Plot PMF and CDF
plot_pmf_cdf(pmf, cdf)
```

```python
from scipy.stats import binom

p = 0.25
N = 100

k1 = 20
prob_exact_20 = binom.pmf(k1, N, p)
print(f"Probability that exactly 20 patients need surgery: {prob_exact_20:.6f}")

k2 = 30
prob_at_least_30 = 1 - binom.cdf(k2 - 1, N, p)
print(f"Probability that at least 30 patients need surgery: {prob_at_least_30:.6f}")

expected_surgeries = N * p
print(f"Expected number of patients requiring surgery: {expected_surgeries}")
```
