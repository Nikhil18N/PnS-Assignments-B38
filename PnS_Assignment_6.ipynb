```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate normally distributed data
mean = 60
std_dev = 12
num_samples = 1000
data = np.random.normal(mean, std_dev, num_samples)

# Plot histogram
plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Plot PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Titles and labels
plt.title("Normal Distribution: Mean=60, Std Dev=12")
plt.xlabel("Data Points")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Given parameters
mean = 30
std_dev = 5

# Probability calculations
p_gte_19 = 1 - stats.norm.cdf(19, mean, std_dev)  # P(Calories >= 19)
p_lte_19 = stats.norm.cdf(19, mean, std_dev)      # P(Calories <= 19)
p_between_19_35 = stats.norm.cdf(35, mean, std_dev) - stats.norm.cdf(19, mean, std_dev)  # P(19 <= Calories <= 35)

# Print probabilities
print(f"P(Calories >= 19): {p_gte_19:.4f}")
print(f"P(Calories <= 19): {p_lte_19:.4f}")
print(f"P(19 <= Calories <= 35): {p_between_19_35:.4f}")

# Visualization
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
y = stats.norm.pdf(x, mean, std_dev)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'b', label='Normal Distribution (Mean=30, Std Dev=5)')
plt.fill_between(x, y, where=(x >= 19), color='g', alpha=0.5, label='P(Calories >= 19)')
plt.fill_between(x, y, where=(x <= 19), color='r', alpha=0.5, label='P(Calories <= 19)')
plt.fill_between(x, y, where=((x >= 19) & (x <= 35)), color='y', alpha=0.5, label='P(19 <= Calories <= 35)')

# Titles and labels
plt.title("Daily Calorie Intake Normal Distribution")
plt.xlabel("Daily Calorie Intake (calories)")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(0)

# Question 1: Generate normal distribution data
mean = 30
std_dev = 7
data_points = np.random.normal(mean, std_dev, 1000)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_points, bins=30, density=True, alpha=0.6, color='g')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mean, std_dev)
plt.plot(x, p, 'k', linewidth=2)
plt.title('Histogram and PDF of Normal Distribution')

# Question 2: Calculate probabilities for heart rate distribution
mean_hr = 50
std_dev_hr = 10

x_hr = np.linspace(0, 100, 1000)
pdf_hr = norm.pdf(x_hr, mean_hr, std_dev_hr)

plt.subplot(1, 2, 2)
plt.plot(x_hr, pdf_hr, 'r', label='PDF of Resting Heart Rate')

prob_30_or_more = 1 - norm.cdf(30, mean_hr, std_dev_hr)
prob_less_than_30 = norm.cdf(30, mean_hr, std_dev_hr)
prob_between_30_and_50 = norm.cdf(50, mean_hr, std_dev_hr) - norm.cdf(30, mean_hr, std_dev_hr)

plt.title('PDF of Resting Heart Rate')
plt.xlabel('Heart Rate (bpm)')
plt.ylabel('Probability Density')
plt.legend()

plt.tight_layout()
plt.show()

print("Probability of RHR >= 30 bpm:", prob_30_or_more)
print("Probability of RHR <= 30 bpm:", prob_less_than_30)
print("Probability of RHR between 30 and 50 bpm:", prob_between_30_and_50)
```
