```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Data
calories = np.array([2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000])
exercise = np.array([60, 50, 30, 90, 45, 20, 75, 95, 15, 10])
bmi = np.array([25, 26, 28, 23, 27, 30, 24, 22, 31, 33])

# Scatter plots
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
sns.scatterplot(x=calories, y=bmi)
plt.xlabel("Daily Calorie Intake")
plt.ylabel("BMI")
plt.title("Calorie Intake vs. BMI")

plt.subplot(1, 3, 2)
sns.scatterplot(x=exercise, y=bmi)
plt.xlabel("Exercise Duration (mins)")
plt.ylabel("BMI")
plt.title("Exercise Duration vs. BMI")

plt.subplot(1, 3, 3)
sns.scatterplot(x=calories, y=exercise)
plt.xlabel("Daily Calorie Intake")
plt.ylabel("Exercise Duration (mins)")
plt.title("Calorie Intake vs. Exercise Duration")

plt.tight_layout()
plt.show()

# Pearson correlation coefficients
corr_calories_bmi, _ = pearsonr(calories, bmi)
corr_exercise_bmi, _ = pearsonr(exercise, bmi)
corr_calories_exercise, _ = pearsonr(calories, exercise)

print(f"Pearson correlation (Calories vs BMI): {corr_calories_bmi:.2f}")
print(f"Pearson correlation (Exercise vs BMI): {corr_exercise_bmi:.2f}")
print(f"Pearson correlation (Calories vs Exercise): {corr_calories_exercise:.2f}")

# Interpretation
if abs(corr_calories_bmi) > 0.5:
    print("There is a strong positive correlation between Calorie Intake and BMI.")
else:
    print("Calorie Intake and BMI have a weak or moderate correlation.")

if abs(corr_exercise_bmi) > 0.5:
    print("There is a strong negative correlation between Exercise Duration and BMI.")
else:
    print("Exercise Duration and BMI have a weak or moderate correlation.")

if abs(corr_calories_exercise) > 0.5:
    print("There is a strong correlation between Calorie Intake and Exercise Duration.")
else:
    print("Calorie Intake and Exercise Duration have a weak correlation.")
```
